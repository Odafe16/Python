# Nigeria Economy Analysis

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\JONATHAN\Desktop\WORK\Nigerian_Economy.csv')
df = df.replace(r'[, %]', '', regex=True)
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
Year = df['Year']
GDP = df['GDP(in bn. US$PPP)']
GDP_growth = df['GDP growth(real)']
Inflation_rate = df['Inflation rate(in Percent)']
Government_debth = df['Government debt(in\xa0% of GDP)']
 
```

```python
plt.plot(Year, GDP, marker="o")

plt.title("GDP the Years")


#plt.xticks(Year, Year.astype(int), rotation=45)
#plt.yticks(df["Inflation_rate(in_Percent)"].values, fontsize=10, rotation=45)



plt.xlabel("Year")
plt.ylabel("GDP in billion($)")


plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
plt.plot(Year, GDP_growth, marker="o")
plt.title("GDP Growth Rate Over the Years")


#plt.xticks(Year, Year.astype(int), rotation=45)
#plt.yticks(df["Inflation_rate(in_Percent)"].values, fontsize=10, rotation=45)



plt.xlabel("Year")
plt.ylabel("GDP growth(%)")


plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
plt.plot(Year, Inflation_rate, marker="o")

plt.title("Inflation Rate Over the Years")


#plt.xticks(Year, Year.astype(int), rotation=45)
#plt.yticks(df["Inflation_rate(in_Percent)"].values, fontsize=10, rotation=45)



plt.xlabel("Year")
plt.ylabel("Inflation Rate(%)")


plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
plt.plot(Year, Government_debth, marker="o")

plt.title("Nigeria Government Debth Over the Years")


#plt.xticks(Year, Year.astype(int), rotation=45)
#plt.yticks(df["Inflation_rate(in_Percent)"].values, fontsize=10, rotation=45)



plt.xlabel("Year")
plt.ylabel("Government Debth(%)")


plt.tight_layout()
plt.grid(True)
plt.show()
```
