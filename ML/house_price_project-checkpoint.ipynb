# House Price Predictor & Loan Risk Classifier
### Built from scratch — following Andrew Ng's ML Specialization (Part 1)

**What we build:**
- **Part A** → Linear Regression to predict house prices
- **Part B** → Logistic Regression to classify loan approval

Everything is implemented from scratch: gradient descent, cost function, feature scaling. No sklearn for the core models.

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import json

np.random.seed(42)
print('Libraries loaded')
````

---
## Step 1 — Generate Synthetic Dataset

We generate 200 houses with 4 features:
- `size_sqft` — house size in square feet
- `num_bedrooms` — number of bedrooms
- `age_years` — age of the house
- `location_score` — neighbourhood score (1–10)

True price formula (with noise):
```
price = 50*size + 10000*bedrooms - 500*age + 8000*location + noise
```

````python
m = 200  # number of training examples

# Raw features
size_sqft      = np.random.randint(600, 4000, m).astype(float)
num_bedrooms   = np.random.randint(1, 6, m).astype(float)
age_years      = np.random.randint(1, 50, m).astype(float)
location_score = np.random.uniform(1, 10, m)

# Ground-truth price
noise = np.random.normal(0, 15000, m)
price = (50 * size_sqft
         + 10000 * num_bedrooms
         - 500  * age_years
         + 8000 * location_score
         + noise)

# Feature matrix X  (m x n)
X_raw = np.column_stack([size_sqft, num_bedrooms, age_years, location_score])
y_price = price

print(f'Dataset shape : X={X_raw.shape}, y={y_price.shape}')
print(f'Price range   : ${y_price.min():,.0f}  –  ${y_price.max():,.0f}')
print(f'\nFirst 3 rows:')
headers = ['size_sqft','bedrooms','age_yrs','loc_score','price($)']
for i in range(3):
    vals = list(X_raw[i]) + [y_price[i]]
    print('  ', dict(zip(headers, [f'{v:.1f}' for v in vals])))
````

---
## Step 2 — Feature Scaling (Z-score Normalisation)

Andrew Ng emphasises feature scaling so gradient descent converges faster.

$$x_{norm} = \frac{x - \mu}{\sigma}$$

We store `mu` and `sigma` so we can scale new inputs at prediction time.

````python
def zscore_normalise(X):
    """Returns normalised X, plus mu and sigma for later use."""
    mu    = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    X_norm = (X - mu) / sigma
    return X_norm, mu, sigma

X_norm, mu, sigma = zscore_normalise(X_raw)

print('Feature means  (mu)   :', np.round(mu, 2))
print('Feature stds   (sigma):', np.round(sigma, 2))
print('Normalised X — mean ≈ 0, std ≈ 1:')
print('  mean:', np.round(X_norm.mean(axis=0), 4))
print('  std :', np.round(X_norm.std(axis=0), 4))
````

## Part A — Linear Regression from Scratch

### 2a. The Model
$$\hat{y} = \vec{w} \cdot \vec{x} + b$$

### 2b. Cost Function (MSE)
$$J(w,b) = \frac{1}{2m} \sum_{i=0}^{m-1} \left( \hat{y}^{(i)} - y^{(i)} \right)^2$$

### 2c. Gradient Descent
$$w_j := w_j - \alpha \frac{\partial J}{\partial w_j} \qquad b := b - \alpha \frac{\partial J}{\partial b}$$

Where the gradients are:
$$\frac{\partial J}{\partial w_j} = \frac{1}{m} \sum_{i=0}^{m-1} \left(\hat{y}^{(i)} - y^{(i)}\right) x_j^{(i)}$$
$$\frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=0}^{m-1} \left(\hat{y}^{(i)} - y^{(i)}\right)$$

````python
# ── Linear Regression core functions ──────────────────────────────────────

def predict_linear(X, w, b):
    """Vectorised prediction: X (m,n), w (n,), b scalar → y_hat (m,)"""
    return X @ w + b


def compute_cost_linear(X, y, w, b):
    """Mean Squared Error cost (divided by 2 as in Andrew Ng's notation)."""
    m = X.shape[0]
    y_hat = predict_linear(X, w, b)
    cost  = np.sum((y_hat - y) ** 2) / (2 * m)
    return cost


def compute_gradient_linear(X, y, w, b):
    """Compute dJ/dw and dJ/db for linear regression."""
    m = X.shape[0]
    y_hat  = predict_linear(X, w, b)
    error  = y_hat - y                  # (m,)
    dj_dw  = (X.T @ error) / m          # (n,)
    dj_db  = np.sum(error) / m          # scalar
    return dj_dw, dj_db


def gradient_descent_linear(X, y, w_init, b_init, alpha, num_iters):
    """Run gradient descent. Returns final w, b and cost history."""
    w = w_init.copy()
    b = b_init
    cost_history = []

    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient_linear(X, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        if i % 100 == 0:
            cost = compute_cost_linear(X, y, w, b)
            cost_history.append(cost)
            if i % 500 == 0:
                print(f'  Iter {i:5d} | Cost: {cost:>15,.2f}')

    return w, b, cost_history


print('Functions defined')
````

````python
# ── Train Linear Regression ────────────────────────────────────────────────
n_features = X_norm.shape[1]
w_init     = np.zeros(n_features)
b_init     = 0.0
alpha      = 0.1
num_iters  = 2000

print('Training linear regression ...')
w_lin, b_lin, cost_hist_lin = gradient_descent_linear(
    X_norm, y_price, w_init, b_init, alpha, num_iters
)

print(f'\nFinal w: {np.round(w_lin, 2)}')
print(f'Final b: {b_lin:,.2f}')
````

````python
# ── Evaluate ───────────────────────────────────────────────────────────────
y_pred_lin = predict_linear(X_norm, w_lin, b_lin)
mae  = np.mean(np.abs(y_pred_lin - y_price))
rmse = np.sqrt(np.mean((y_pred_lin - y_price) ** 2))
ss_res = np.sum((y_price - y_pred_lin) ** 2)
ss_tot = np.sum((y_price - np.mean(y_price)) ** 2)
r2   = 1 - ss_res / ss_tot

print(f'MAE  : ${mae:>10,.0f}')
print(f'RMSE : ${rmse:>10,.0f}')
print(f'R²   : {r2:.4f}   (1.0 = perfect)')
````

````python
# ── Plot: Cost curve + Predicted vs Actual ─────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Linear Regression Results', fontsize=15, fontweight='bold')

# Cost curve
ax = axes[0]
ax.plot(np.arange(len(cost_hist_lin)) * 100, cost_hist_lin, color='#2563EB', lw=2)
ax.set_xlabel('Iteration')
ax.set_ylabel('Cost J(w,b)')
ax.set_title('Cost Curve — Gradient Descent')
ax.grid(True, alpha=0.3)

# Predicted vs Actual
ax = axes[1]
ax.scatter(y_price / 1000, y_pred_lin / 1000, alpha=0.5, s=20, color='#2563EB')
lims = [min(y_price.min(), y_pred_lin.min()) / 1000,
        max(y_price.max(), y_pred_lin.max()) / 1000]
ax.plot(lims, lims, 'r--', lw=1.5, label='Perfect prediction')
ax.set_xlabel('Actual Price ($k)')
ax.set_ylabel('Predicted Price ($k)')
ax.set_title(f'Predicted vs Actual  (R² = {r2:.3f})')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("C:/Users/JONATHAN/Downloads/python/Data/Coursere Andrew NG Machine Learning/Supervised Machine Learning, Classification and Regression", dpi=150, bbox_inches='tight')
plt.show()
print('Plot saved')
````

## Part B — Logistic Regression (Loan Approval)

Now we classify whether a buyer gets **loan approved** based on:
- Their predicted house price (from Part A)
- Their annual income
- Their credit score

**Rule**: approved if `price / (income * 5) < 1` AND `credit_score > 600` (with noise)

### The Sigmoid function
$$g(z) = \frac{1}{1 + e^{-z}}$$

### Logistic cost (Binary Cross-Entropy)
$$J(w,b) = -\frac{1}{m} \sum_{i=0}^{m-1} \left[ y^{(i)} \log(\hat{y}^{(i)}) + (1 - y^{(i)}) \log(1 - \hat{y}^{(i)}) \right]$$

````python
# ── Generate loan dataset ──────────────────────────────────────────────────
annual_income  = np.random.randint(30000, 200000, m).astype(float)
credit_score   = np.random.randint(450, 850, m).astype(float)

# Debt-to-income ratio based on predicted price
dti = y_pred_lin / (annual_income * 5)   # >1 means house costs more than 5× income

# Label: 1 = approved, 0 = denied
approval_noise = np.random.uniform(0, 1, m)
y_loan = ((dti < 1.0) & (credit_score > 600) | (approval_noise > 0.85)).astype(float)
# Add some misclassification noise
flip_mask = np.random.rand(m) < 0.05
y_loan[flip_mask] = 1 - y_loan[flip_mask]

print(f'Approved : {int(y_loan.sum())} / {m}  ({100*y_loan.mean():.1f}%)')
print(f'Denied   : {int((1-y_loan).sum())} / {m}')

# Feature matrix for logistic regression
X_loan_raw  = np.column_stack([y_pred_lin, annual_income, credit_score])
X_loan_norm, mu_loan, sigma_loan = zscore_normalise(X_loan_raw)
````

````python
# ── Logistic Regression core functions ────────────────────────────────────

def sigmoid(z):
    """σ(z) = 1 / (1 + e^{-z})"""
    return 1.0 / (1.0 + np.exp(-z))


def predict_logistic(X, w, b):
    """Returns probabilities P(y=1 | x)."""
    return sigmoid(X @ w + b)


def compute_cost_logistic(X, y, w, b):
    """Binary cross-entropy loss."""
    m = X.shape[0]
    y_hat = predict_logistic(X, w, b)
    # Clip to avoid log(0)
    y_hat = np.clip(y_hat, 1e-15, 1 - 1e-15)
    cost  = -np.mean(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
    return cost


def compute_gradient_logistic(X, y, w, b):
    """Gradients for logistic regression — same formula as linear!"""
    m = X.shape[0]
    y_hat = predict_logistic(X, w, b)
    error = y_hat - y
    dj_dw = (X.T @ error) / m
    dj_db = np.sum(error) / m
    return dj_dw, dj_db


def gradient_descent_logistic(X, y, w_init, b_init, alpha, num_iters):
    w = w_init.copy()
    b = b_init
    cost_history = []

    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient_logistic(X, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        if i % 100 == 0:
            cost = compute_cost_logistic(X, y, w, b)
            cost_history.append(cost)
            if i % 500 == 0:
                print(f'  Iter {i:5d} | Cost: {cost:.6f}')

    return w, b, cost_history


print('Logistic functions defined')
````

````python
# ── Train Logistic Regression ──────────────────────────────────────────────
n_loan     = X_loan_norm.shape[1]
w_init_log = np.zeros(n_loan)
b_init_log = 0.0

print('Training logistic regression ...')
w_log, b_log, cost_hist_log = gradient_descent_logistic(
    X_loan_norm, y_loan, w_init_log, b_init_log,
    alpha=0.1, num_iters=2000
)

print(f'\nFinal w: {np.round(w_log, 4)}')
print(f'Final b: {b_log:.4f}')
````

````python
# ── Evaluate Logistic Regression ───────────────────────────────────────────
probs      = predict_logistic(X_loan_norm, w_log, b_log)
y_pred_log = (probs >= 0.5).astype(int)

accuracy  = np.mean(y_pred_log == y_loan)
tp = np.sum((y_pred_log == 1) & (y_loan == 1))
tn = np.sum((y_pred_log == 0) & (y_loan == 0))
fp = np.sum((y_pred_log == 1) & (y_loan == 0))
fn = np.sum((y_pred_log == 0) & (y_loan == 1))
precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall    = tp / (tp + fn) if (tp + fn) > 0 else 0

print(f'Accuracy  : {accuracy*100:.1f}%')
print(f'Precision : {precision*100:.1f}%')
print(f'Recall    : {recall*100:.1f}%')
print(f'\nConfusion matrix:')
print(f'  TP={tp}  FP={fp}')
print(f'  FN={fn}  TN={tn}')
````

````python
# ── Plot: Cost curve + Decision boundary (2 features) ─────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Part B — Logistic Regression Results', fontsize=15, fontweight='bold')

# Cost curve
ax = axes[0]
ax.plot(np.arange(len(cost_hist_log)) * 100, cost_hist_log, color='#16A34A', lw=2)
ax.set_xlabel('Iteration')
ax.set_ylabel('Cost J(w,b)')
ax.set_title('Cost Curve — Logistic Gradient Descent')
ax.grid(True, alpha=0.3)

# Decision boundary using features 0 (price) and 2 (credit score)
ax = axes[1]
x0_vals = X_loan_norm[:, 0]
x2_vals = X_loan_norm[:, 2]

approved = y_loan == 1
ax.scatter(x0_vals[approved],  x2_vals[approved],  alpha=0.6, s=25, color='#16A34A', label='Approved')
ax.scatter(x0_vals[~approved], x2_vals[~approved], alpha=0.6, s=25, color='#DC2626', label='Denied', marker='x')

# Decision boundary: w0*x0 + w1*x1_mean + w2*x2 + b = 0  → solve for x2
x0_range = np.linspace(x0_vals.min(), x0_vals.max(), 100)
w1_x1_mean = w_log[1] * np.mean(X_loan_norm[:, 1])  # hold x1 at mean
x2_boundary = -(w_log[0] * x0_range + w1_x1_mean + b_log) / (w_log[2] + 1e-10)
ax.plot(x0_range, x2_boundary, 'k--', lw=2, label='Decision boundary')

ax.set_xlabel('Predicted Price (normalised)')
ax.set_ylabel('Credit Score (normalised)')
ax.set_title(f'Decision Boundary  (Accuracy = {accuracy*100:.1f}%)')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("C:/Users/JONATHAN/Downloads/python/Data/Coursere Andrew NG Machine Learning/Supervised Machine Learning, Classification and Regression", dpi=150, bbox_inches='tight')
plt.show()
print('Plot saved')
````

## Step 4 — Make a Prediction (Manual Test)

Let's manually predict for a new house + buyer.

````python
# ── New house: 1800 sqft, 3 bedrooms, 10 years old, location 7/10 ──────────
new_house = np.array([1800, 3, 10, 7.0])
new_house_norm = (new_house - mu) / sigma
predicted_price = predict_linear(new_house_norm, w_lin, b_lin)

print(f'House features : size={new_house[0]} sqft, {int(new_house[1])} beds, {int(new_house[2])} yrs old, loc={new_house[3]}/10')
print(f'Predicted price: ${predicted_price:,.0f}')

# ── Buyer: income $75k, credit score 720 ──────────────────────────────────
buyer_income = 75000.0
buyer_credit = 720.0

loan_input = np.array([predicted_price, buyer_income, buyer_credit])
loan_input_norm = (loan_input - mu_loan) / sigma_loan
approval_prob   = predict_logistic(loan_input_norm, w_log, b_log)
decision        = 'APPROVED' if approval_prob >= 0.5 else 'DENIED'

print(f'\nBuyer: income=${buyer_income:,.0f}, credit={int(buyer_credit)}')
print(f'Loan approval probability: {approval_prob*100:.1f}%')
print(f'Loan decision: {decision}')
````
